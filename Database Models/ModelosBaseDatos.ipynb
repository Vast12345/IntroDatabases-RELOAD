### **Plain Archives**

Types of Plain Archives:
* CSV
* Marcados (HTML)
* Sin Formato
* Binary Archives
Characteristics:
*Cons*
* Structures barely organized
* Small volumes of data
* No security
* Difficult to relate
* Easy to corrupt
*Pros*
* Easy to compress
* Easy to transfer Data
* Formats humanly understandable

### **Modelos Jerarquico**

Characteristics:
* Inverted Tree
* 3 Levels (Raiz, Ramas, Hojas)
* Sumamente Simple
* Muy Rigido
* No permite relaciones N:M

### **Modelo Relacional**

**The Vision for a relational Database**

* Los datos deben describirse en su estructura natural
* Provee las bases para un lenguaje de consulta de alto nivel
* Sienta unas bases estables para el tratamiento     de la Derivabilidad, Redundancia y Consistencia de las Relaciones.
* Permite evaluar claramente al ambito, las limitaciones logicas y las bondades de los Sistemas de Informacion.

*Una tabla en un modelo relacional esta contribuido de atributos(columnas) y tuplas(filas). Cada tupla es un conjunto y cada relacion es un conjunto. Cambiando el orden de los elementos en el conjunto NO define un nuevo conjunto.*

[TablaRelacional](/home/camper/IntroDatabases-RELOAD/content/TablaRelacion.png)

**Operaciones sobre conjuntos**:
* Union: A U B
* Interseccion: A n B
* Diferencia: A\B = A - B
* COmplementado ¬A = A^c
* Diferencia simetrica: A △ B
* Producto cartesiano: A x B

*Las Bases - Teoria de Conjuntos -> Algebra de Conjuntos*

**Modelo Relacional Informacion**
* Creado por Edgar F. Codd(1969)
* Basado en la Teoria de Conjuntos
* Va mas alla de la definicion de un Modelo de Datos
* Lenguaje basado en la Logica de Predicados
* Tratamiento de la Derivabilidad, Redudencia y Consistencia de las Relaciones
* Evaluacion de Limitaciones y Bondades de los Sistemas de Informacion
